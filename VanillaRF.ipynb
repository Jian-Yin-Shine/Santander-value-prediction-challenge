```
import pandas as pd
import numpy as np
```

```
pd.options.display.max_columns = 900
```

```
train_df = pd.read_csv('train.csv', low_memory = False)
test_df = pd.read_csv('test.csv', low_memory= False)
```

```
train_df.head()
```

```
train_Id = train_df['ID']
```

```
y_train = train_df['target']
```

```
train_y = np.log1p(y_train.values) 
```

```
train_df.drop('ID', axis =1, inplace = True)
```

```
train_df.drop('target', axis = 1,inplace = True)
```

```
train_df.head()
```

```
from sklearn.ensemble import RandomForestRegressor
```

```
m = RandomForestRegressor(n_estimators = 100, n_jobs = -1, max_features= 0.5)
m.fit(train_df, train_y)
```

```
test_df.head()
```

```
test_ID = test_df['ID']
```

```
test_df.drop('ID', axis = 1, inplace = True)
```

```
test_df.head()
```

```
prediction = m.predict(test_df)
```

```
Submit = pd.DataFrame({'ID': test_ID, 'target': prediction})
Submit.to_csv('VanillaRf(log).csv', index = False)
```
